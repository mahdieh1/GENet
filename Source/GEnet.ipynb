```python
import requests
import os
import zipfile

def read_urls_from_file(file_path):
    with open(file_path, 'r') as file:
        urls = file.readlines()
    return [url.strip() for url in urls]

def download_tsv(urls, destination_folder):
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Limit to first 500 URLs
    urls = urls[:500]
    downloaded_files = []

    for url in urls:
        if url.endswith('.tsv'):
            file_name = url.split('/')[-1]
            destination_path = os.path.join(destination_folder, file_name)
            try:
                response = requests.get(url)
                if response.status_code == 200:
                    with open(destination_path, 'wb') as file:
                        file.write(response.content)
                        downloaded_files.append(destination_path)
                    print(f"Downloaded {file_name}")
                else:
                    print(f"Failed to download {file_name}. Status code: {response.status_code}")
            except Exception as e:
                print(f"Error occurred while downloading {file_name}: {e}")
    return downloaded_files

def zip_files(files, destination_folder, zip_name="downloaded_files.zip"):
    zip_path = os.path.join(destination_folder, zip_name)
    with zipfile.ZipFile(zip_path, 'w') as zipf:
        for file in files:
            zipf.write(file, os.path.basename(file))
    print(f"Created ZIP file at: {zip_path}")

# Usage
text_file_path = '/content/geneexpression.txt'  # Replace with the path to your text file
destination_folder = '/content/'  # Replace with the path where you want to save the TSV files and ZIP file

urls = read_urls_from_file(text_file_path)
downloaded_files = download_tsv(urls, destination_folder)
zip_files(downloaded_files, destination_folder)
```

```python
import pandas as pd
import zipfile
import os
from pathlib import Path

# Function to unzip and load peak data
def unzip_and_load_peak_data(zip_path, sample, data_type, temp_dir):
    # Make sure the temp directory exists
    if not os.path.exists(temp_dir):
        os.makedirs(temp_dir)

    # Attempt to find and extract the specific BED file
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        expected_file_name = f"{sample}_{data_type}.bed"
        try:
            zip_ref.extract(expected_file_name, temp_dir)
        except KeyError:
            print(f"File {expected_file_name} not found in {zip_path}.")
            return None  # File not found

    # Load the extracted BED file
    file_path = os.path.join(temp_dir, expected_file_name)
    if os.path.exists(file_path):
        peak_data = pd.read_csv(file_path, sep='\t', header=None, names=['chr', 'start', 'end'])
        os.remove(file_path)  # Cleanup
        return peak_data
    else:
        print(f"Extracted file {expected_file_name} not found after extraction.")
        return None
```

```python
import pandas as pd
import gzip
import zipfile
import os

def unzip_and_decompress_peak_data(zip_path, sample, temp_dir):
    # Adjusted to expect .bed.gz files directly named after the sample without _TF or _H3k27ac suffix
    expected_file_name_gz = f"{sample}.bed.gz"
    extracted_file_path = os.path.join(temp_dir, expected_file_name_gz)

    # Extract the .bed.gz file from the zip
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        try:
            zip_ref.extract(expected_file_name_gz, temp_dir)
        except KeyError:
            print(f"File {expected_file_name_gz} not found in {zip_path}.")
            return None  # File not found

    # Decompress the .bed.gz file
    decompressed_file_path = extracted_file_path[:-3]  # Remove .gz extension
    with gzip.open(extracted_file_path, 'rb') as f_in:
        with open(decompressed_file_path, 'wb') as f_out:
            f_out.write(f_in.read())

    # Now, you can load the decompressed BED file
    if os.path.exists(decompressed_file_path):
        peak_data = pd.read_csv(decompressed_file_path, sep='\t', header=None, names=['chr', 'start', 'end'])
        os.remove(extracted_file_path)  # Cleanup .bed.gz file
        os.remove(decompressed_file_path)  # Cleanup decompressed .bed file
        return peak_data
    else:
        print(f"Decompressed file {decompressed_file_path} not found after decompression.")
        return None
```

```python
tf_zip_path = '/content/TF.zip'  # Path to the ZIP file containing TF peak data
h3k27ac_zip_path = '/content/H3k27ac.zip'  # Path to the ZIP file containing H3k27ac peak data
temp_dir = "/content/temp_peak_data"  # Temporary directory for extracted files

tf_peak_data = {}
h3k27ac_peak_data = {}

for sample in common_names:
    # Process TF peak data
    tf_peak_data[sample] = unzip_and_decompress_peak_data(tf_zip_path, sample, temp_dir)
    # Process H3k27ac peak data
    h3k27ac_peak_data[sample] = unzip_and_decompress_peak_data(h3k27ac_zip_path, sample, temp_dir)
```

```python
import pandas as pd
from pathlib import Path

# Initialize an empty DataFrame to hold aggregated gene expression data
gene_expression_df = pd.DataFrame()

# Iterate over each TSV file in the specified directory
for tsv_file in Path(extraction_dir).glob('*.tsv'):
    # Read the current TSV file into a DataFrame
    df = pd.read_csv(tsv_file, sep='\t')

    # Dynamically identify the gene_id column, falling back to target_id if necessary
    gene_id_col = next((col for col in df.columns if 'gene_id' in col.lower()), None)
    if not gene_id_col:
        gene_id_col = next((col for col in df.columns if 'target_id' in col.lower()), 'target_id')

    # Identify the TPM column
    tpm_col = next((col for col in df.columns if 'tpm' in col.lower()), None)

    # Verify that necessary columns were found
    if not tpm_col:
        print(f"No TPM data found in {tsv_file.name}.")
        continue

    # Use TPM as the expression value
    df['expression_value'] = df[tpm_col]

    # Use the stem of the file name as the cell line identifier
    df['cell_line'] = tsv_file.stem

    # Select and rename columns to standardize the DataFrame structure
    df = df[[gene_id_col, 'expression_value', 'cell_line']].rename(columns={gene_id_col: 'gene_id'})

    # Aggregate the current DataFrame into the collective gene_expression_df
    gene_expression_df = pd.concat([gene_expression_df, df], ignore_index=True)

# Display the first few rows of the aggregated gene expression DataFrame
print(gene_expression_df.head())
```

```python
def check_peak_overlap(peak_data, tss, chr, start_pos, end_pos):
    """Check if there's an overlap between a peak and a given TSS region."""
    overlaps = peak_data[(peak_data['chr'] == chr) & (peak_data['end'] >= start_pos) & (peak_data['start'] <= end_pos)]
    return not overlaps.empty

def create_feature_matrix(genes_df, peak_data_dict, position_range=2000):
    """Create a binary feature matrix indicating peak overlaps around TSS for each gene."""
    # Initialize the feature matrix with zeros
    index = pd.MultiIndex.from_product([genes_df['gene_id'], ['TF', 'H3k27ac']], names=['gene_id', 'feature_type'])
    columns = [f"pos_{i}" for i in range(-position_range, position_range + 1)]
    feature_matrix = pd.DataFrame(0, index=index, columns=columns)

    # Iterate through each gene to update the feature matrix
    for _, gene_row in genes_df.iterrows():
        gene_id, tss, chr = gene_row['gene_id'], gene_row['tss'], gene_row['chr']
        for sample, peaks in peak_data_dict.items():
            for pos in range(-position_range, position_range + 1):
                start_pos, end_pos = tss + pos - 1, tss + pos + 1  # Define a small window to check for overlap
                for feature_type in ['TF', 'H3k27ac']:
                    if check_peak_overlap(peaks, tss, chr, start_pos, end_pos):
                        feature_matrix.at[(gene_id, feature_type), f"pos_{pos}"] = 1

    return feature_matrix
```
